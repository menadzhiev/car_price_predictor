# Предобработка данных

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('cars_df.csv')
print(df.shape)
df.head()
```

Подготовленный нами датасет представляет из себя 2982 объявления, опубликованных на американском сайте cars.com, специализирующемся на вторичном рынке автомобилей. Каждое объявление содержит информацию о продаваемой подержанной машине и ее конечную или же желаемую продавцом стоимость.

### Анализ признаков 

Итак, предоставленный для анализа датасет имеет 15 признаков/колонок и 1 целевую переменную, которую мы в финальной части нашего проекта будем предсказывать. Это переменная 'Price' - то есть цена машины. Теперь давайте пройдемся по каждому столбцу нашей таблицы и посмотрим, что в них содержится:
   * **Model** - название модели машины, включающее в себя так же год выпуска и марку автомобиля 
   * **href** - атрибут адреса документа, в нашем случае объявления
   * **Exterior color** - цвет машины 
   * **Interior color** - цвет салона автомобиля 
   * **Drivetrain** - трансмиссия автомобиля, т.е. информация о его ведущие колеса 
   * **MPG** - количество миль, которое автомобиль сможет проехать на одном галлоне топлива
   * **Fuel type** - тип топлива 
   * **Transmission** - информация о коробке передач
   * **Engine** - информация о двигатели автомобиля, содержащая 
      * объем двигателя 
      * конфигурацию 
      * количество клапанов
      * дополнительные характеристки двигателя
   * **VIN** - уникальный идентификационный номер транспортного средства 
   * **Stock #** - складской номер автомобиля
   * **Mileage** - пробег 
   * **Accidents or damage** - наличие аварий у машины 
   * **1-owner vehicle** - информация о количестве владельцев автомобиля, один ли владелец у машины 
   * **Personal use only** - предназначенность машины исключительно для личных целей
   
   
Посмотрев на содержание нашей исходной, еще не обработанной таблицы, мы можем сразу отметить, что такие признаки, как href, VIN и Stock # имеют уникальную для каждой машины информацию и поэтому при дальнейшем построении модели и проверке гипотез они окажутся слишком сложны в анализе и не принесут никакой ценности, поэтому от них стоит сразу избавиться. 

Отметим еще такой признак, как  Transmission, характеризующий коробку передач автомобиля. Запись этого признака каждым из владельцев оказывается индивидуальной, что делает сложным дальнейшую обработку этого столбца и его кодирование в целом.Более того, обльшинство машин на американском рынке имеют автоматическую коробку передач, поэтому исключение из расмотрения этой категории приведет к незначительным изменениям в построении модели и проверке гипотез. Таким образом, как и опысанные выше признаки, мы его удалим.

Так же нетрудно заметить, что некоторые столбцы представляют из себя набор сразу из нескольких признаков, например, Model или Engine. Такие колонки требуют дальнейшей обработки, ведь их компоненты оказываются достаточно значимимы при определении цены машины на вторичном рынке. Поэтому давайте перейдем к их обработке.


```python
# избавимся от слишком уникальных признаков
df.drop(columns=['VIN', 'Stock #', 'href', 'Transmission'], inplace=True)
print(df.shape)
df.head()
```

Порой в нашей таблице встречаются пропуски, предтсавляющие из себя не заполненную продавцами информацию о той или иной характеристке машины. Проанализируем их количество в зависимости от признака:

```python
df.isna().sum() #пропуски в данных отсутствуют
```

Однако в наших данных есть иная проблема. Многие продавцы решают по определенным причинам ставить вместо информации значок пропуска, сиволизирующий либо о нежелании распространять информацию, либо же ее отсутствие, а может быть и вовсе пропуск из-за невнимательности. В таких категориальных признаках признаках, как тип топлива и транссмисия, их следует убрать, ведь заменить их нельзя будет, так как мы не знаем какие столь важные технические характериски подойдут именно этой машине.

```python
df = df.drop(df[df['Fuel type'] == '–'].index)
df = df.drop(df[df['Drivetrain'] == '–'].index)
```

### Обработка колонок

#### Обработка будущих вещественных признаков и целевой переменной

Начнем предобработку данных с преобразования будущих вещественных признаком и самой целевой переменной, а именно приведе их к числовому виду. Такими колонками являются Mileage, Price, MPG.


```python
# определим функцию для цены
def new_price(price):
    return float(price.replace('$', '').replace(',', ''))

df['Price'] = df['Price'].apply(new_price)
```

```python
df = df.drop(df[df['Mileage'] == 'None reported'].index) # избавимся от аномальных значений, содержащихся в колонке
df = df.drop(df[df['Mileage'] == 'At least 1 accident or damage reported'].index)
```

```python
# определим функцию для пробега
def new_mileage(mileage):
    return float(mileage.replace('mi.', '').replace(',', ''))

df['Mileage'] = df['Mileage'].apply(new_mileage)
```

После преобразования цены автомобиля и ее пробега, давайте посмотрим на колонку MPG. Как было описано ранее, данный признак представляет из себя удельный израсход топлива при минимальном и максимальном использовании мощностей автомобиля (нижний и верхний орог интервала соотвественно). Из этого признака мы сможем получить сразу три: MPG_min - минимальный израсход топлива,  MPG_max - максимальный и  MPG_mean - средний израсход топлива автомобилем.

```python
hlp = df['MPG'].str.split('\n\n\n\n', expand=True)  # выделим интервал израсхода топлива от дополнительной информации 
hlp.rename(columns={0: 'MPG_par', 1: 'Data'}, inplace=True)
data = pd.concat([df, hlp], axis=1)
data.drop('MPG', axis=1, inplace=True)
```

```python
data.head()  # записали интервал в промежуточную колонку MPG_par, а дополнительную информацию в Data
```

```python
k = data['MPG_par'].str.split('–', expand=True)  # отделим теперь границы интервала израсхода топлива
k.rename(columns={0: 'MPG_min', 1: 'MPG_max'}, inplace=True)
data = pd.concat([data,k], axis=1)
data.drop(columns=['MPG_par', 'Data'], inplace=True)
```

```python
data.head()
```

Нетрудно заметить, что у нас появились попуски в данных, которые можно заполнить средними показателями по каждому из столбцов:

```python
data.fillna('', inplace=True)  # в ходе преобразований получились наны, уберем их
```

```python
usl_empty = data['MPG_min'] == ''  # определим условия 
usl_empty_1 = data['MPG_max'] == ''
```

```python
data['MPG_min'][usl_empty] = 0  # вставим в пустые строки нули, чтобы преобразовать в дальнейшем столбцы в вещественные
data['MPG_max'][usl_empty_1] = 0
```

```python
data['MPG_min'] = data['MPG_min'].astype(float)
data['MPG_max'] = data['MPG_max'].astype(float)
```

```python
usl_0 = data['MPG_min'] == 0  # заполним средним значением нулевое
a = data['MPG_min'][~usl_0].mean()
data['MPG_min'][usl_0] = a
```

```python
usl_0 = data['MPG_max'] == 0  # заполним средним значением нулевое
a = data['MPG_max'][~usl_0].mean()
data['MPG_max'][usl_0] = a
```

```python
data['MPG_mean'] = (data['MPG_min'] + data['MPG_max']) / 2
```

```python
data.head()  # посмотрим, что у нас получилось
```

### Обработка будущих категориальных признаков

Посмотрев на такие переменные, как Drivetrain, Exterior color и Interior color, мы можем заметить, что одна и та же информация может быть записана по-разному. В случае трансмиссии это полные и короткие названия типов привода. А вот с цветами солона и машины дело обстоит интереснее. Оба этих признака принимают достаточно большое количество значений. Однако исключение подобных категориальных признаков будет не совсем уместным,ведь цвет играет порой определяющую роль в стоимости машины, поэтому предлагается выделить основные/базовые цвета автомобилей и относить к ним их различные оттенки, с целью упрощения дальнейшего анализа. Если оттенок определить сложно или же будет возникать пропущенное значение в данных, то будет относить такие объекты к одельному типу цвета - Other, то есть другое.

```python
data['Drivetrain'].unique()  # посмотрим, какие значения принимает признак
```

```python
usl_f = data['Drivetrain'] == 'FWD'  # приведем все значение к полному виду
data.loc[usl_f,'Drivetrain' ] = 'Front-wheel Drive'

usl_a = data['Drivetrain'] == 'AWD'
data.loc[usl_a,'Drivetrain' ] = 'All-wheel Drive'

usl_4 = data['Drivetrain'] == '4WD'
data.loc[usl_4,'Drivetrain' ] = 'Four-wheel Drive'

usl_r = data['Drivetrain'] == 'RWD'
data.loc[usl_r,'Drivetrain' ] = 'Rear-wheel Drive'
```

```python
# Определим необходимую функцию для работы с цветами
def color_detecktor(color):
    if "Black" in color or "Ebony" in color:
        return "Black"
    elif "White" in color:
        return "White"
    elif "Silver" in color or "Gray" in color:
        return "Silver/Gray"
    elif "Red" in color:
        return "Red"
    elif "Blue" in color:
        return "Blue"
    elif "Green" in color:
        return "Green"
    elif "Yellow" in color:
        return "Yellow"
    elif "Orange" in color:
        return "Orange"
    elif "Brown" in color or "Carmelo" in color or "Tan" in color:
        return "Brown"
    elif "Gold" in color:
        return "Gold"
    elif "Metallic" in color:
        return "Metallic"
    return "Other"
```

```python
# Обработаем цвета
data['Exterior color'] = data['Exterior color'].apply(color_detecktor)
data['Interior color'] = data['Interior color'].apply(color_detecktor)
```

### Обработка составных признаков

Теперь займемся разбиение переменных на новые составляющие:

```python
model = df['Model'].str.split(' ', expand=True)  # достанем из столбца модель год выпуска и марку машины
model.rename(columns={0: 'Year', 1: 'Brand'}, inplace=True) 
data = pd.concat([data,model.iloc[:,:2]], axis=1)
data.drop('Model', axis=1, inplace=True)
```

```python
data['Brand'].unique()  # посмотрим, какие фирмы у нас присутствуют 
```

```python
data.loc[data['Brand'] == 'Land', 'Brand'] = 'Land Rover' # запишем полные названия некоторых марок
data.loc[data['Brand'] == 'Aston', 'Brand'] = 'Aston Martin'
data.loc[data['Brand'] == 'Alfa', 'Brand'] = 'Alfa Romeo'
```

```python
data['Year'] = data['Year'].astype(int)  # преобразуем переменную year в число
```

```python
data.head()
```

```python
# Вновь определим необходимую функцию для двигателя
def arr_engine_mask(arr):
    if len(arr) == 3:
        if (arr[0][-1] == 'L') & (arr[2][-1] == 'V') & ((arr[1][0] == 'I') | (arr[1][0] == 'V')):
            return arr
    return None
```

```python
# Обработаем двигателя
engine_str = data['Engine']

engine_arr = engine_str.str.split()
engine_arr = engine_arr.apply(lambda x: x[:3])


engine_df = engine_arr.apply(arr_engine_mask).to_frame()
engine_df = engine_df.rename(columns={'Engine': 'engine_arr'})


cars_df = pd.concat([data, engine_df], join='inner', axis=1)
cars_df.dropna(inplace=True)
```

```python
# Выделим признаки
capacity = cars_df['engine_arr'].apply(lambda x: x[0])
configuration = cars_df['engine_arr'].apply(lambda x: x[1])
valves = cars_df['engine_arr'].apply(lambda x: x[2])

engine_df = pd.concat(
    [capacity.rename('Capacity'), configuration.rename('Configuration'), valves.rename('Valves')],
    join='inner', axis=1
)
```

```python
cars_df = pd.concat([cars_df, engine_df], join='inner', axis=1)
cars_df.drop(columns='engine_arr', inplace=True)
cars_df.drop(columns=['Engine'], inplace=True)
```

```python
# Определим функцию для литров, чтобы преобразовать в число объем двигателя
def new_capacity(capacity):
    return float(capacity.replace('L', ''))

cars_df['Capacity'] = cars_df['Capacity'].apply(new_capacity)
```

```python
# Определим функцию для количества клапанов
def new_valves(valves):
    return float(valves.replace('V', ''))

cars_df['Valves'] = cars_df['Valves'].apply(new_valves)
```

```python
cars_df.head()
```

```python
# Перезапишем столбцы
new_order = ['Brand', 'Year', 'Exterior color', 'Interior color',
             'Drivetrain', 'Fuel type', 'Capacity', 'Configuration', 
             'Valves', 'Accidents or damage', '1-owner vehicle',
            'Personal use only', 'MPG_min', 'MPG_max', 'MPG_mean', 'Mileage', 'Price']

cars_df = cars_df.reindex(columns=new_order)
print(cars_df.shape)
cars_df.head()
```

Таким образом, у нас получилась итоговая таблица, содержащая следующие признаки:

   * **MPG_max** - количество миль, которое автомобиль сможет проехать на одном галлоне топлива при максимальном использовании *(вещественный признак)*

И одна целевая переменная **Price** - цена автомобиля, доллары.
